# enumerate() function genetates  the pairs of (index, value) for each item

# for index ,value in enumerate(iterable)

```python
# 1. accessing items using enumerate()
```

```python
cars=["Aston","Audi","Mclaren"]
for i,x in enumerate(cars):                     #index=i
    print(x)
```

```python
#2. accessing items and indexes enumerate()
```

```python
cars=["Aston","Audi","Mclaren"]
for x in enumerate(cars,start=1):                     
    print(x[0],x[1])                 #print(index,value)
```

# why x[0] and x[1]?

In the code for x in enumerate(cars): print(x[1]), the expression x[1] is used to print only the value component of the pair generated by enumerate(), which is the second element of the tuple.
 When enumerate() is used with a list like cars, it generates pairs of (index, value) for each item in the list. In this case, each x value in the loop represents a pair, where x[0] is the index of the current item in the list, and x[1] is the value of the current item in the list.

Therefore, when the expression x[1] is used in the print() statement, it prints only the value component of the current pair, which is the value of the current item in the cars list.

#So, the output of the code for x in enumerate(cars): print(x[1]) would be:
this is because the loop iterates over the cars list, and for each item in the list, it generates a pair containing the index and the value of the item. The print() statement then prints only the value component of each pair, which is the corresponding car brand name.
